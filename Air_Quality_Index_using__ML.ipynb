**Forecasting using FB Prophet algorithm**

```python
#converting date format DD/MM/YYYY to YYYY-MM-DD

date_info = pd.to_datetime(new_air_quality_data['Date'], format='%d/%m/%Y')
print(date_info)
```

```python
#combining 2 series data to dataframe

date_time = pd.concat([date_info, time_info], axis=1)
```

```python
date_time['ds'] = date_time['Date'].astype(str) +' '+ date_time['Time'].astype(str)
```

```python
date_time.info()
```

We have to convert ds from object datatype to date time format

```python
data = pd.DataFrame()
```

```python
data['ds'] = pd.to_datetime(date_time['ds'], format='%Y-%m-%d %H.%M.%S')
```

```python
data.head()
```

```python
data['y'] = new_air_quality_data['RH']
```

```python
data.head()
```

```python
from prophet import Prophet
```

```python
# train-test split (last 20% for testing)
train_size = int(len(data) * 0.8)

train = data.iloc[:train_size]
test = data.iloc[train_size:]
```

```python
#training the prophet model

model = Prophet()
model.fit(train)
```

```python
future = model.make_future_dataframe(
    periods=len(test),
    freq='H',
    include_history=False
)

forecast = model.predict(future)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

y_true = test['y'].values
y_pred = forecast['yhat'].values

mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))

print("MAE:", mae)
print("RMSE:", rmse)
```

```python
mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
print("MAPE:", mape)
```
